```python
def grade(per: int)->str:
    if per >= 80:
        return "A+"
    elif per >= 70:
        return "A"
    elif per >= 60:
        return "B"
    elif per >= 50:
        return "C"
    elif per >= 40:
        return "D"
    elif per >= 33:
        return "E"
    else:
        return "F"

grade(82)
```

```python
# df['percentage'] # gives percentaage
df_grade = df['percentage'].apply(grade) # run grade function on percentage in a loop
df_grade
```
